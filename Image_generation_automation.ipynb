```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
import random

# Website URL and credentials
website_url = "https://piclumen.com/app/account"
username = "zeeshanali12512@gmail.com"  # Replace with your email
password = "Happy12412"  # Replace with your password

# List of prompts
inputt_prompt = [ 
    "A high-energy aerial view of a packed stadium during the Super Bowl, with the field illuminated by bright lights and fans creating a sea of team colors.",  
    "A close-up of the Vince Lombardi Trophy being held high by the winning team, with confetti raining down and players celebrating in the background.",  
    "A dramatic shot of the halftime show, with a world-famous artist performing on a massive stage, surrounded by pyrotechnics and a roaring crowd.",  
    "A panoramic view of the Super Bowl stadium at night, with the iconic logo projected onto the field and fireworks lighting up the sky.",  
    "A close-up of a quarterback throwing a perfect spiral pass, with the ball in mid-air and the crowd holding their breath in anticipation.",  
    "A behind-the-scenes shot of the Super Bowl broadcast team, with cameras, monitors, and crew members capturing every moment of the game.",  
    "A fan’s perspective of the Super Bowl, with a sea of jerseys, foam fingers, and face paint, all cheering passionately for their team.",  
    "A close-up of the Super Bowl ring, with intricate details and gemstones sparkling under the stadium lights.",  
    "A dramatic shot of a game-winning touchdown, with a player diving into the end zone and the crowd erupting in celebration.",  
    "A bird’s-eye view of the Super Bowl halftime show stage, with elaborate props, dancers, and special effects creating a visual spectacle.",  
    "A close-up of the Super Bowl logo painted on the field, with players running past it during a crucial moment in the game.",  
    "A shot of the Super Bowl trophy presentation, with the winning team’s coach and players hoisting the Vince Lombardi Trophy amidst a shower of confetti.",  
    "A fan’s view of the Super Bowl tailgate party, with grills, team flags, and fans enjoying food and drinks before the big game.",  
    "A close-up of a referee making a decisive call, with players and coaches reacting passionately on the sidelines.",  
    "A dramatic shot of a field goal attempt, with the ball in mid-air and the crowd holding their breath as it sails toward the goalposts.",  
    "A behind-the-scenes shot of the Super Bowl halftime show rehearsal, with performers, choreographers, and technicians preparing for the big moment.",  
    "A close-up of a player’s helmet, with the team logo and scratches from the game, symbolizing the intensity of the Super Bowl.",  
    "A panoramic view of the Super Bowl stadium during the national anthem, with a flyover by military jets and the crowd holding their hands over their hearts.",  
    "A close-up of a fan holding a handmade sign, with a heartfelt message or clever pun, cheering passionately for their team.",  
    "A dramatic shot of the Super Bowl MVP celebrating on the field, with teammates lifting them onto their shoulders and the crowd chanting their name."  
]  

# Set up the Chrome WebDriver
driver = webdriver.Chrome()

try:
    # Open the website
    driver.get(website_url)

    # Wait for the email input field
    email_field = WebDriverWait(driver, 15).until(
        EC.visibility_of_element_located((By.CSS_SELECTOR, "input[type='email']"))
    )

    # Log in to the website
    email_field.send_keys(username)
    password_field = driver.find_element(By.CSS_SELECTOR, "input[type='password']")
    password_field.send_keys(password)

    sign_in_button = WebDriverWait(driver, 15).until(
    EC.presence_of_element_located((By.XPATH, "/html/body/div[1]/div/div[1]/div[2]/div[1]/div/div[3]/div/div[4]/button"))
    )
    sign_in_button.click()

    # Wait until the dashboard page is loaded
    WebDriverWait(driver, 15).until(
        EC.presence_of_element_located((By.XPATH, "/html/body/div/div/div[1]/div[1]/div[1]/div/div[1]/div/nav"))
    )

    # Click the button to access the prompt section
    button_xpath = "/html/body/div/div/div[1]/div[1]/div[1]/div/div[1]/div/nav/div[1]/div[2]/div[2]/a/div"
    button = WebDriverWait(driver, 15).until(
        EC.element_to_be_clickable((By.XPATH, button_xpath))
    )
    driver.execute_script("arguments[0].scrollIntoView();", button)
    button.click()

    # Wait for the page to load
    WebDriverWait(driver, 5).until(
        EC.presence_of_element_located((By.XPATH, "//*[@id='app']/div/div[1]/div[2]/div/div[1]/div/div[1]"))
    )

    # Iterate through prompts
    for prompt_text in inputt_prompt:
        try:
            # Locate textarea
            textarea_xpath = "//*[@id='app']/div/div[1]/div[2]/div/div[1]/div/div[1]/div[1]/div/div/div/div[1]/textarea[1]"
            textarea = WebDriverWait(driver, 15).until(
                EC.visibility_of_element_located((By.XPATH, textarea_xpath))
            )
            textarea.clear()  # Clear existing text
            textarea.send_keys(prompt_text)

            # Simulate Ctrl + Enter keypress
            actions = ActionChains(driver)
            actions.key_down(Keys.CONTROL).send_keys(Keys.ENTER).key_up(Keys.CONTROL).perform()

            # Wait before processing next prompt
            wait_time = random.randint(60, 70)
            print(f"Waiting for {wait_time} seconds before processing the next prompt...")
            time.sleep(wait_time)

        except Exception as e:
            print(f"Error processing prompt: {prompt_text} - {str(e)}")

finally:
    driver.quit()  # Close the browser when done
```
